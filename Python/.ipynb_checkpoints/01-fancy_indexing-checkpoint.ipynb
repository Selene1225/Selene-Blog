比较如下两种可以达到同样效果的写法：

```python
import numpy as np
row = [0,1,2]
indices = [1,4,0]
arr = np.zeros((len(row), 5), dtype=np.float32)

print(arr)
```

```python
# Option 1: Using for loop
for i in row:
    col = indices[i]
    arr[i, col] = 1.0

print("update by for loop:\n", arr)
```

```python
# Option 2: Using fancy indexing
arr[row, indices] = 1.0
print("update by fancy indexing:\n", arr)
```

np.arange 作用是生成一个按照指定步长排列的等差序列ndarray数组，无需提前计算序列长度，直接根据边界和步长生成数组。

- Option 2-FancyIndexing是瞬移：cpu/gpu 看到两个数组，直接并行计算内存偏移量，一次性完成赋值。
- Option 2-For是慢动作：每次迭代都要进行类型检查、内存寻址，效率极低。
